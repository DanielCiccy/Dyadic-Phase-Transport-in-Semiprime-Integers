```python
import numpy as np
import pandas as pd
import gmpy2
from gmpy2 import mpfr

TAU = 2*np.pi
gmpy2.get_context().precision = 200

def wrap_pi(x):
    return (x + np.pi) % (2*np.pi) - np.pi

def wrap_0_2pi(x):
    return x % (2*np.pi)

def unit(v):
    n = np.hypot(v[:,0], v[:,1])
    return v / n[:,None]

def perp(v):
    return np.column_stack([-v[:,1], v[:,0]])

def angle_between(u, v):
    cross = u[:,0]*v[:,1] - u[:,1]*v[:,0]
    dot   = u[:,0]*v[:,0] + u[:,1]*v[:,1]
    return np.arctan2(cross, dot)

def theta_tangent_pure_from_hex(n_hex: str) -> float:
    n_int = int(n_hex, 16)
    lx = gmpy2.log2(mpfr(n_int))
    u = float(lx - int(lx))
    return TAU * u

def R_of_k(k, k_ref=31):
    return 2.0 ** (k - k_ref)

df = pd.read_csv("semiprimes_64_500k_enriched.csv")

# angles de placement (Ï†n reconstruit, Ï†p fourni)
phi_n = np.array([theta_tangent_pure_from_hex(x) for x in df["n"]], dtype=float)
phi_p = df["theta_pmin"].to_numpy(dtype=float)

k_n = df["k_n"].to_numpy(dtype=int)
k_p = df["k_pmin"].to_numpy(dtype=int)

Rn = R_of_k(k_n, k_ref=31)
Rp = R_of_k(k_p, k_ref=31)

Pn = np.column_stack([Rn*np.cos(phi_n), Rn*np.sin(phi_n)])
Pp = np.column_stack([Rp*np.cos(phi_p), Rp*np.sin(phi_p)])

v_np = Pp - Pn
u_np_perp = unit(perp(v_np))
u_rad_n   = unit(Pn)   # rayon en n

theta_geom = wrap_0_2pi(angle_between(u_rad_n, u_np_perp))

theta_csv = df["theta_n"].to_numpy(dtype=float)

diff = wrap_pi(theta_csv - theta_geom)

print("âŸ¨|Î¸_csv âˆ’ Î¸_geom|âŸ© =", np.mean(np.abs(diff)))
print("Ïƒ(Î¸_csv âˆ’ Î¸_geom) =", np.std(diff))
```

```python
import numpy as np
import pandas as pd
import gmpy2
from gmpy2 import mpfr

TAU = 2*np.pi
gmpy2.get_context().precision = 200

def wrap_pi(x):
    return (x + np.pi) % (2*np.pi) - np.pi

def wrap_0_2pi(x):
    return x % (2*np.pi)

def hex_to_int(s):
    s = s.strip().lower()
    return int(s, 16) if s.startswith("0x") else int(s)

def theta_tangent_pure_int(x_int: int) -> float:
    lx = gmpy2.log2(mpfr(int(x_int)))
    u = float(lx - int(lx))
    return TAU * u

df = pd.read_csv("semiprimes_64_500k_enriched.csv")

# parse p,q
p = df["p"].map(hex_to_int).to_numpy(dtype=object)
q = df["q"].map(hex_to_int).to_numpy(dtype=object)
pmin = np.minimum(p, q)
qmax = np.maximum(p, q)

theta_n = df["theta_n"].to_numpy(float)
theta_pmin = df["theta_pmin"].to_numpy(float)

# reconstruire theta_qmax (tangente pure)
theta_qmax = np.array([theta_tangent_pure_int(x) for x in qmax], float)

theta_star = wrap_0_2pi(theta_pmin + theta_qmax)

diff = wrap_pi(theta_n - theta_star)

print("âŸ¨|Î¸n - (Î¸pmin+Î¸qmax)|âŸ© =", np.mean(np.abs(diff)))
print("Ïƒ(diff) =", np.std(diff))
print("median(|diff|) =", np.median(np.abs(diff)))
print("p90(|diff|) =", np.quantile(np.abs(diff), 0.90))
```
